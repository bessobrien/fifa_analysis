```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
# Import provisional files
matches = "data_source/international_matches.csv"
players = "data_source/players_22.csv"
```

```python
# Create data frames
matches_df = pd.read_csv(matches, index_col=0)
players_df = pd.read_csv(players, index_col=0)
```

```python
# Check players df
players_df.head()
```

Use SQLAlechamy to Import DataFrames into Postgres

```python
# Create database connection string
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/fifa_analysis"
```

```python
# Create the database engine
engine = create_engine(db_string)
```

```python
# Import dataframes to database
matches_df.to_sql(name='matches', con=engine)
players_df.to_sql(name='players', con=engine)
```
